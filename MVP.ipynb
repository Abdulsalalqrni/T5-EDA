```python
low_avg=df.low.median()
low_avg
```

```python
no_trades_avg=df["no_trades "].median()
no_trades_avg
```

# Q1) What the maximum opening in the Saudi Stock Market(Tadawul)?

```python
df["open"].max()
```

```python
plt.figure(figsize =[40,25],dpi=250);
plt.plot(df["open"])
plt.title("The maximum opening in the Saudi Stock Market(Tadawul)",fontsize =40)
plt.ylabel("The opening price",fontsize =40)
plt.rc('xtick',labelsize=35) 
plt.rc('ytick',labelsize=35)
plt.savefig("photo.jpg")
```

# Q2) What is the lowest and highest performance day in the Saudi Stock Market(Tadawul)?

```python
lowest_day=df.groupby("day")[["low"]].max()
lowest_day
```

```python
highest_day=df.groupby("day")[["high"]].max()
highest_day
```

```python
plt.figure(figsize=[27,11],dpi=250)
plt.suptitle("The lowest and highest performance day in the Saudi Stock Market",fontsize =21)
plt.subplot(1,2,1)
plt.barh(lowest_day.index,lowest_day["low"],color="#2e8b57")
plt.title("The lowest performance day in the Saudi Stock Market",fontsize =19)
plt.xlabel("The lowest price of the stock at that day",fontsize =19);
plt.ylabel("The day of the stock price",fontsize =19);
plt.rc('xtick',labelsize=19) 
plt.rc('ytick',labelsize=19)
plt.subplot(1,2,2)
plt.barh(highest_day.index,highest_day["high"],color="#8b2e2e")
plt.title("The highest performance day in the Saudi Stock Market",fontsize =19);
plt.xlabel("The highest price of the stock at that day",fontsize =19);
plt.ylabel("The day of the stock price",fontsize =19);
plt.rc('xtick',labelsize=19) 
plt.rc('ytick',labelsize=19)
plt.savefig("image.jpg")
```

# Q3) What are the top five companies in the Saudi Stock Market(Tadawul)?

```python
top_5_co=df.groupby("name")[["high"]].max().reset_index().sort_values("high",ascending=False).head()
top_5_co
```

```python
plt.figure(figsize = [6,6])
sns.barplot(x=top_5_co["name"],y=top_5_co["high"])
plt.title(" Top five companies in the Saudi Stock Market",fontsize =15);
plt.xlabel("Name of the company",fontsize =15);
plt.ylabel("The highest price of the stock at that day",fontsize =15);
plt.rc('xtick',labelsize=12) 
plt.rc('ytick',labelsize=11)
plt.xticks(rotation=90)
plt.grid()
plt.savefig("figure.jpg")
```
